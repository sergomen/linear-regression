```python
print(st.pearsonr(data['Weight'], data['Length1']))
print(st.pearsonr(data['Weight'], data['Length2']))
print(st.pearsonr(data['Weight'], data['Length3']))
print(st.pearsonr(data['Weight'], data['Height']))
print(st.pearsonr(data['Weight'], data['Width']))
print(st.pearsonr(data['Length1'], data['Length2']))
print(st.pearsonr(data['Length1'], data['Length3']))
print(st.pearsonr(data['Length1'], data['Height']))
print(st.pearsonr(data['Length1'], data['Width']))
print(st.pearsonr(data['Length2'], data['Length3']))
print(st.pearsonr(data['Length2'], data['Height']))
print(st.pearsonr(data['Length2'], data['Width']))
print(st.pearsonr(data['Length3'], data['Height']))
print(st.pearsonr(data['Length3'], data['Width']))
print(st.pearsonr(data['Height'], data['Width']))
```

```python
pole1='Length3'
pole2='Height'
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
z=np.polyfit(data[pole1], data[pole2], deg=1)
f = np.poly1d(z)
plt.scatter(data[pole1], data[pole2], s=2)
plt.scatter(data[pole1], f(data[pole1]), s=1)
print(f)
residuals = data[pole2]-f(data[pole1])
plt.subplot(1,2,2)
plt.scatter(data[pole1], residuals, s=2)
print(np.var(residuals))
```

```python
y=data['Length3'].values
x=data[['Width', 'Height']].values.T
def reg_m(y, x):
	ones = np.ones(x.shape[1])
	X = sm.add_constant(np.column_stack((x[0], x[1], ones)))
	results = sm.OLS(y,X).fit()
	return results
print(reg_m(y,x).summary())
```
